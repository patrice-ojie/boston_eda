```python
import pandas as pd
import numpy as np
import seaborn as sns                       
import matplotlib.pyplot as plt             
%matplotlib inline     
sns.set(color_codes=True)
```

```python
# Name the columns, load the dataset and display the first 5 rows of the dataset
column_names = ['Crime Rate', 'Residential Land Zones', 'Proportion of Non-Retail Businesses', 'Charles River Boundary', 'Nitric Oxide Concentration', 
                'Number of Rooms', 'Proportion of Units Pre-1940', 'Distance to Boston Employment Centre', 'Accessibility to Radial Highways', 
                'Tax Rate per $10000', 'Pupil-Teacher Ratio', 'B', 'Percentage of Lower Status', 'Median Value in $1000s']
data = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
data.head()  
```

```python
# Display the bottom 5 rows of the dataset
data.tail()
```

```python
# Check the types of data within the dataset
data.dtypes
```

```python
# Drop columns that are not of use in this EDA and display the first 5 columns again
data = data.drop(['Residential Land Zones', 'Charles River Boundary','B'], axis=1)
data.head(5)
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
data.shape
```

```python
# Heat Map to determine dependent variables
plt.figure(figsize=(10,5))
c= data.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c 
```

```python
# Histogram
sns.histplot(x=data['Number of Rooms'])
plt.title("Average Number of Rooms per Dwelling")
plt.ylabel('Frequency')
plt.xlabel('Number of Rooms')
```

```python
# Scatterplot
sns.scatterplot(x=data['Median Value in $1000s'], y=data['Crime Rate'])
plt.xlabel('Median Value in $1000s')
plt.ylabel('Crime Rate')
```
